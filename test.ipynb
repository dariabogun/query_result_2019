# Задача
Задача звучит так - создать функцию на языке Python, которая для указанного произвольного пользователя берет его историю шагов и на выходе  отдает список размеченных дней: неизвестно, обычный или необычный.

Ваша цель - максимально попасть в субъективное ощущение пользователя, чтобы он согласился с вашим мнением о том, что сегодня он прошел 10 000 шагов и это необычно. Или сегодня он прошел 6 000 шагов и в этом нет ничего необычного.

```python
df.columns=['TimeEnd', 'TimeEndLocalTz', 'TimeOffset', 'TimeStart', 'TimeStartLocalTz', 'ID', 'Value'] #переименование колонок для удобства
print(df)
```

```python
#изменение формата даты, т.к. для работы нужен только день (время не важно)
def set_new_time(df, dfColumn):
    df[dfColumn] = pd.to_datetime(df[dfColumn])
    df[dfColumn] = df[dfColumn].map(lambda x: x.strftime('%Y-%d-%m'))

set_new_time(df,'TimeEnd')
set_new_time(df,'TimeEndLocalTz')
set_new_time(df,'TimeStart')
set_new_time(df,'TimeStartLocalTz')

df.head(20)
```

```python
import difflib
#нахождение разницы между столбцами, содержащие дату
i=0

while i!=len(df):
    matcher = difflib.SequenceMatcher(None, df.TimeEnd[i], df.TimeStart[i])
    if matcher.ratio()!=1:
        print(matcher.ratio())
    i+=1
#разницы нет, все колонки с датами после изменения формата полностью идентичны
#это означает, что все наблюдения были проведены в рамках одних суток
```

```python
df.drop(['TimeEndLocalTz', 'TimeOffset', 'TimeStart', 'TimeStartLocalTz', 'ID'], axis='columns', inplace=True)  #убираем лишние столбцы
df.groupby(df.TimeEnd).count() #группируем по дате шаги
df.head(20)
```

```python
dn = df.groupby(['TimeEnd']).sum() #суммируем шаги по дате
dn.head(5)
```

```python
pd.DataFrame(dn.Value.describe())
```

```python
dn['Type'] = str(np.nan)
```

```python
import warnings 
warnings.filterwarnings("ignore") 
i=0
while i!=len(dn):
    if dn.Value[i]<dn.Value.mean()-dn.Value.std()or dn.Value[i]>dn.Value.mean()+dn.Value.std():
        dn.Type[i] = 'необычный'
    elif dn.Value.mean()-dn.Value.std()<=dn.Value[i]<=dn.Value.mean()+dn.Value.std():
        dn.Type[i] = 'обычный' 
    else:
        dn.Type[i] = 'неизвестно' 
    i+=1 
dn.head(30)
```

# Итог:
Мы задали нижнюю и верхнюю границы для определения диапазона "нормальности" путем нахождения среднего значения и стандартного отклонения значения Value. Таким образом данный диапазон будет подобран индивидуально под каждого пользователя.
